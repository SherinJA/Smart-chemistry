```python
#21PT27 - SARMITHA . V.D
#21PT28 - SHERIN. J. A


#LEXICAL ANALYSER FOR CHEMICAL EQUATIONS

from prettytable import PrettyTable
table=PrettyTable(["Token", "Token type"])

def tokenize_chemical_formula(formula):
    tokens = []
    i = 0
    while i < len(formula):
        char = formula[i]

        # Skip spaces
        if char.isspace():
            i += 1
            continue

        # Check for operators or special characters
        if char in '+→()':
            token_type = {'+': 'PLUS', '\u2192': 'ARROW', '(': 'LPAREN', ')': 'RPAREN'}.get(char)
            tokens.append((token_type, char, i))
            i += 1
            continue

        # Check for digits (coefficients or subscripts)
        if char.isdigit():
            start = i
            while i < len(formula) and formula[i].isdigit():
                i += 1
            value = formula[start:i]
            # Determine if it's a coefficient or subscript based on the previous token
            if tokens and (tokens[-1][0] == 'ELEMENT' or tokens[-1][0] == 'RPAREN'):
                token_type = 'SUBSCRIPT'
            else:
                token_type = 'COEFFICIENT'
            tokens.append((token_type, value, start))
            continue

        # Check for elements
        if char.isalpha():
            start = i
            i += 1
            # Capture two-character element symbols
            if i < len(formula) and formula[i].islower():
                i += 1
            element = formula[start:i]
            tokens.append(('ELEMENT', element, start))
            continue

        # If none of the above, skip the character
        i += 1

    return tokens

# Test input
data = 'H2O + NaCl + (NH4)2SO4 \u2192 Na2SO4 + 2NH3 + H2O'

# Tokenize
tokens = tokenize_chemical_formula(data)

# Create a PrettyTable
table = PrettyTable()
table.field_names = ["Token Type", "Token Value", "Position"]

# Fill table with data
for token_type, value, pos in tokens:
    table.add_row([token_type, value, pos])

# Print the table
print(table)
```
